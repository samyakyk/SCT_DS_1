```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()
filename = list(uploaded.keys())[0]

df = pd.read_csv(filename, skiprows=4)

df = df.drop(columns=["Unnamed: 69"], errors="ignore")

latest_year = "2023"
df_latest = df[["Country Name", latest_year]].dropna()

top10 = df_latest.sort_values(by=latest_year, ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top10["Country Name"], top10[latest_year])
plt.xticks(rotation=45, ha="right")
plt.title("Top 10 Countries by Population (2023)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.show()

plt.figure(figsize=(8,5))
plt.hist(df_latest[latest_year], bins=20)
plt.title("Distribution of Population Across Countries (2023)")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.show()
```
